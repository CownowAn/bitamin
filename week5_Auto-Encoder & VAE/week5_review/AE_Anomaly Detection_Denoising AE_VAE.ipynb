# < AutoEncoder >

# 1. 개념

1. Encoder, Decoder가 각각 담당하는 부분은?

- Encoder는 차원 축소 역할을 하고, Decoder는 생성 모델의 역할을 하는 경우가 많습니다.

2. Autoencoder의 활용 예시


(1)   Stacking AutoEncoder for pre-training
  - weight값을 초기화할 때 사용되기도 했었습니다.


  (2)   Classification
  - AE를 이용해 학습시킨 Encoder를 Future extractor로 사용하여 self-supervised learnig 또는 semi-supervised learning에 활용하기도 합니다.


  (3)   Clustering
  - 비슷한 unlabled data간에는 숨겨진 row-dimensional latent representation이 존재할 것이라고 가정합니다. 이 가정을 바탕으로, AE를 이용해 학습시킨 Encoder를 통해 unlabled data를 embedding합니다. 그리고 embedding된 관측치를 이용해 기존의 clustering 알고리즘을 수행합니다.


  (4)   Anomaly Detection
  - 우선, 정상 관측치와 이상 관측치 간에 latent structure가 다를 것이라고 가정합니다. 따라서, 정상 관측치만을 학습데이터로 사용함으로써 AE가 이상 관측치는 잘 복원하지 못할 것이라고 가정합니다. 즉, 검증과정에서 이상 관측치가 입력되면 reconstruction loss가 클 것이라고 예상하는 것입니다.


# 2. Anomaly Detection

autoencoder는 manifold learning과 generator로 구성되어 활용성이 매우 많습니다. 그 중에서 autoencoder가 이상치 탐지에 쓰이는 사례를 살펴보겠습니다.

## 데이터에 대한 설명
ECG5000 데이터세트를 사용하겠습니다. 이 데이터는 약 5000개의 개별적인 심전도로 이루어져있고, 각 관측치마다 약 140개의 data point가 있습니다

(행 5000, 열 140).



이 데이터세트는 마지막 열에 anomalous 여부가 있기 때문에 supervised learning problem이기도 합니다. y값은 0(비정상 리듬에 해당) 또는 1(정상 리듬에 해당)으로 레이블이 지정됩니다.

## Autoencoder을 활용하여 이상 탐지하기
autoencoder를 사용해서 어떻게 이상 관측치를 분류할 수 있을까요? 가장 중요한 것은 저희는 훈련으로 정상 리듬만을 훈련 시킬 것이라는 점입니다. 이 때, 저희가 배웠듯이 autoencoder는 reconstruction error를 최소화하도록 훈련됩니다.

다음으로 test로 이상 관측치를 돌릴 것인데, 이 때 reconstruction error가 정해준 임계값보다 크다고 생각을 합니다. 즉, 이상 관측치와 정상관측치는 다른 latent structure을 갖고 있다고 생각하기 때문입니다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, losses
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
```

데이터를 받아오겠습니다.

```python
dataframe = pd.read_csv('http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv', header=None)
raw_data = dataframe.values
dataframe.head()
```

```python
# 행이 대략 5000, 열 레이블까지 포함하여 141
dataframe.shape
```

```python
# 레이블 생성하기
labels = raw_data[:, -1]

# 나머지 데이터 가져오기
data = raw_data[:, 0:-1]
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, test_size=0.2, random_state=21
)
```

```python
# 데이터 정규화
min_val = tf.reduce_min(train_data)
max_val = tf.reduce_max(train_data)

train_data = (train_data - min_val) / (max_val - min_val)
test_data = (test_data - min_val) / (max_val - min_val)

train_data = tf.cast(train_data, tf.float32)
test_data = tf.cast(test_data, tf.float32)
```

```python
# 훈련시킬 때 정상 관측치만 활용할 것이기 때문에 두 종류의 관측치를 구분해주는 것이 필요
train_labels = train_labels.astype(bool)
test_labels = test_labels.astype(bool)

normal_train_data = train_data[train_labels]
normal_test_data = test_data[test_labels]

anomalous_train_data = train_data[~train_labels]
anomalous_test_data = test_data[~test_labels]
```

```python
# 정상적인 심전도는 어떻게 보이는지 시각화
plt.grid()
plt.plot(np.arange(140), normal_train_data[0])
plt.title("A Normal ECG")
plt.show()
```

```python
# 이상 관측치 시각화
plt.grid()
plt.plot(np.arange(140), anomalous_train_data[2])
plt.title("An Anomalous ECG")
plt.show()
```

## Autoencoder 생성

## 2.1. autoencoder 구조는 140 - 32 - 16 - 8(latent space) - 16 - 32 - 140(output)으로 만들 것입니다. 

(은닉층의 활성화 함수는 모두 relu)

```python
encoder = keras.models.Sequential([
                                           layers.Dense(units=140, input_dim=140, activation='relu'),
                                           layers.Dense(units=32, activation='relu'),
                                           layers.Dense(units=16, activation='relu'),
                                           layers.Dense(units=8)
])

decoder = keras.models.Sequential([
                                           layers.Dense(units=16, input_dim=8, activation='relu'),
                                           layers.Dense(units=32, activation='relu'),
                                           layers.Dense(units=140)
])

ae = keras.models.Sequential([encoder,decoder])
```

## 2.2. autoencoder 모델에서 optimzer은 adam, loss는 mae로 하겠습니다.

```python
ae.compile(optimizer='adam', loss='mae')
```

## 2.3. 학습을 진행합니다. 단, 학습 데이터 중 정상 데이터만 학습시켜야합니다.

```python
# 참고로, autoencoder가 정상데이터는 매우 잘 학습하지만, 
# 이상치가 섞인 데이터는 이보다 더딘 것을 확인하기 위해서
# validation data는 test_data(정상,이상 관측치)를 넣습니다.

history = ae.fit(normal_train_data, normal_train_data, 
          epochs=20, 
          batch_size=512,
          validation_data=(test_data, test_data),
          shuffle=True)
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
```

```python
# 정상인 데이터를 한 경우 error을 시각화해서 확인
encoded_imgs = encoder(normal_test_data).numpy()
decoded_imgs = decoder(encoded_imgs).numpy()

# 파란색은 input, 빨간색은 ouput으로 사이에 있는 핑크색 영역이 크면 loss가 큰 것
plt.plot(normal_test_data[0], 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], normal_test_data[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
# 비정상인 데이터를 시각화한 경우
# 확실히 정상인 데이터에 대한 reconstruction이 input을 따라가지 못하는 것 확인 가능
encoded_imgs = encoder(anomalous_test_data).numpy()
decoded_imgs = decoder(encoded_imgs).numpy()

plt.plot(anomalous_test_data[2], 'b')
plt.plot(decoded_imgs[2], 'r')
plt.fill_between(np.arange(140), decoded_imgs[2], anomalous_test_data[2], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
# 이상 감지를 하기 위해서는 임계값 필요
# 훈련 세트에서 정상 예제에 대한 평균 오차를 계산하여서 이를 바탕으로 임계값 고정
reconstructions = ae.predict(normal_train_data)
train_loss = tf.keras.losses.mae(reconstructions, normal_train_data)

plt.hist(train_loss, bins=50)
plt.xlabel("Train loss")
plt.ylabel("No of examples")
plt.show()
```

```python
print(np.std(train_loss))
```

```python
reconstructions = ae.predict(anomalous_test_data)
test_loss = tf.keras.losses.mae(reconstructions, anomalous_test_data)

plt.hist(test_loss, bins=50)
plt.xlabel("Test loss")
plt.ylabel("No of examples")
plt.show()
```

## 2.4. 원하는 임계값을 numpy.float32 형식으로 변환해줍니다.

```python
# 데이터 loss의 평균에 표준편차를 더한 값으로 임계값 설정
threshold = np.mean(train_loss) + np.std(train_loss)

print("Threshold: ", threshold)
```

```python
# 정한 임계값에 따라서 loss가 그보다 클 경우 이상 데이터로 판단

def predict(model, data, threshold):
  reconstructions = model(data)
  loss = tf.keras.losses.mae(reconstructions, data)
  return tf.math.less(loss, threshold)

def print_stats(predictions, labels):
  print("정확도 = {}".format(accuracy_score(labels, preds)))
  print("정밀도 = {}".format(precision_score(labels, preds)))
  print("재현율 = {}".format(recall_score(labels, preds)))
```

```python
preds = predict(ae, test_data, threshold)
print_stats(preds, test_labels)
```

## 3.2. convolutional AE 모델을 형성합니다.

```python
input_img = Input(shape=(28, 28, 1))

#encoder 부분
x = Conv2D(32, (3,3), activation='relu',padding='same')(input_img)    #input: 28*28*1 -> output: 28*28*32
x = MaxPooling2D((2,2), padding='same')(x)                           #input: 28*28*32 -> output: 14*14*32
x = Conv2D(64, (3,3), activation='relu', padding='same')(x)           #input: 14*14*32 -> output: 14*14*64
encoded = MaxPooling2D((2,2), padding='same')(x)                     #input: 14*14*64 -> output: 7*7*64


#decoder 부분 
x = Conv2D(64, (3,3), activation='relu',padding='same')(encoded) #input: 7*7*64 -> output: 7*7*64
x = UpSampling2D((2,2))(x)                                       #input: 7*7*64 -> output: 14*14*64                                        
x = Conv2D(32, (3,3), activation='relu', padding='same')(x)      #input: 14*14*64 -> output: 14*14*32
x = UpSampling2D((2,2))(x)                                       #input: 14*14*32 -> output: 28*28*32 
decoded = Conv2D(1, (3,3), activation='sigmoid', padding='same')(x) #input: 28*28*32 -> output: 28*28*1

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

encoder = Model(input_img, encoded)

#모델 시각화를 통해 구조 파악
plot_model(autoencoder, 
           to_file='AE.png',
           show_shapes=True)
```

아래는 VAE의 Encoding Network를 함수로 만든 코드입니다.

```python
# Encoder 
input_encoder = (28, 28, 1)                                                                                         # 인코더의 input shape

def encoder(input_encoder):
    
    inputs = keras.Input(shape=input_encoder, name='input_layer')                                                   # input자료의 shape 넣어주기

    x = layers.Conv2D(32, kernel_size=3, strides= 2, padding='same', activation='relu',name='conv_1')(inputs)       # 첫 번째 Conv layer
    
    x = layers.Conv2D(64, kernel_size=3, strides= 2, padding='same', activation='relu', name='conv_2')(x)           # 두 번째 Conv layer
    
    x = layers.Flatten()(x)
                                                                                                                    # Conv layer의 결과물을 flatten시킨 후
    x = layers.Dense(32, activation='relu')(x)                                                                      # Nerual Network 통과
    
    mean = layers.Dense(2, name='mean')(x)                                                                          # 잠재변수 z의 평균, shape=(None, 2)
    log_var = layers.Dense(2, name='log_var')(x)                                                                    # 잠재변수 z의 log 분산, shape=(None, 2)
    model = tf.keras.Model(inputs, (mean, log_var), name="Encoder")                                                 # encoder 모델의 ouput은 (mean, log_var)
    return model
```

```python
# train 과정에서 좀 더 용이하게 사용하기 위해 enc로 지정하기
enc = encoder(input_encoder)
```

아래는 VAE의 Decoding Network를 함수로 만든 코드입니다. 아래 출력한 그림과 결과가 동일하게 나오도록 빈 칸을 채워주세요!

```python
# Decoder
input_decoder = (2,)                                                                                                     # 디코더의 input shape

def decoder(input_decoder):
    
    inputs = keras.Input(shape=input_decoder, name='input_layer')                                                        # latent variable z의 shape 넣어주기

    x = layers.Dense(3136, name='dense_1')(inputs)                                                                       # encoder의 마지막 Conv layer의 노드수만큼 유닛 지정
    x = layers.Reshape((7, 7, 64), name='Reshape_Layer')(x)                                                              # Conv2DTranspose를 통과할 수 있는 모양으로 reshape
  
    x = layers.Conv2DTranspose(64, 3, strides= 2, padding='same', activation='relu', name='conv_transpose_1')(x)         # 첫 번째 ConvTranspose layer

    x = layers.Conv2DTranspose(32, 3, strides= 2, padding='same', activation='relu', name='conv_transpose_2')(x)         # 두 번째 ConvTranspose layer

    outputs = layers.Conv2DTranspose(1, 3, 1, padding='same', activation='sigmoid', name='conv_transpose_3')(x)           # 마지막은 유닛 개수 1개


    model = tf.keras.Model(inputs, outputs, name="Decoder")
    return model
```

```python
# train 과정에서 좀 더 용이하게 사용하기 위해 dec로 지정하기
dec = decoder(input_decoder)
```

## VAE - Sampling 설정

```python
# 밑에 sampling 함수 내에서 사용할 함수를 설정
def sampling_model(distribution_params):                                                                # encoder 모델의 output인 (mean, log_var)을 input으로 받음
    mean, log_var = distribution_params
    epsilon = K.random_normal(shape=K.shape(mean), mean=0., stddev=1.)                                  # backpropagation이 가능하도록 reparamatization trick 사용
    return mean + K.exp(log_var / 2) * epsilon


# z를 sampling하는 함수 sampling을 설정
def sampling(distribution_params):                                                                      # encoder 모델의 output인 (mean, log_var)을 input으로 받음
    mean, log_var = distribution_params
    z=layers.Lambda(sampling_model, output_shape=(2,), name='encoder_output')([mean, log_var])          # Lambda로 은닉층 추가하여 z값 sampling함
    return z                                                                                            # sampling한 z값 반환                              
```

## VAE - train 과정에서 사용할 함수들 설정하기
진행 과정에서 유사한 함수를 추정해 사용하기 때문에 아래과 같이 임의로 지정하는 과정이 필요합니다.

```python
def mse_loss(y_true, y_pred):                                                                           # reconstruction loss 계산
    r_loss = K.mean(K.square(y_true - y_pred), axis = [1,2,3])
    return 1000 * r_loss


def kl_loss_custom(mean, log_var):                                                                      # Kullback_Leibler loss 계산
    kl_loss_custom =  -0.5 * K.sum(1 + log_var - K.square(mean) - K.exp(log_var), axis = 1)
    return kl_loss_custom                                                                               #이름을 kl_loss로 하면 원래 있는 변수와 충돌. 다른 이름으로 바꿔주기


def vae_loss(y_true, y_pred, mean, log_var):                                                            # 총 loss 계산
    r_loss = mse_loss(y_true, y_pred)
    global kl_loss
    kl_loss = kl_loss_custom(mean, log_var)
    return  r_loss + kl_loss
```

```python
def train_step(images):

    with tf.GradientTape() as encoder, tf.GradientTape() as decoder:
      
        mean, log_var = enc(images, training=True)                                      # encoder에서 z의 평균과 log 분산 반환
        latent = sampling([mean, log_var])                                              # z 생성
        generated_images = dec(latent, training=True)                                   # decoder로 새로운 이미지 생성
        loss = vae_loss(images, generated_images, mean, log_var)                        # 총 loss 계산

        
    gradients_of_enc = encoder.gradient(loss, enc.trainable_variables)                  # encoder에서 훈련 가능한 모수들에 대해 미분
    gradients_of_dec = decoder.gradient(loss, dec.trainable_variables)                  # decoder에서 훈련 가능한 모수들에 대해 미분
    
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0005)                        # Adam optimizer 사용

    optimizer.apply_gradients(zip(gradients_of_enc, enc.trainable_variables))
    optimizer.apply_gradients(zip(gradients_of_dec, dec.trainable_variables))
    return loss
```

```python
os.makedirs('tf_vae/fashion/training_weights', exist_ok=True)
os.makedirs('tf_vae/fashion/images', exist_ok=True)
```

```python
def train(dataset, epochs):
    for epoch in range(epochs):
        start = time.time()
        i = 0
        loss_ = []
        for image_batch in dataset:
            i += 1
            loss = train_step(image_batch)
            #loss_.append(loss)

        #print("Loss",np.mean(loss_))    
        seed = image_batch[:25]
        display.clear_output(wait=True)
        generate_and_save_images([enc,sampling,dec],
                                  epoch + 1,
                                  seed)
        # Save the model every 15 epochs
        #if (epoch + 1) % 15 == 0:
        #checkpoint.save(file_prefix = checkpoint_prefix)
        enc.save_weights('tf_vae/fashion/training_weights/enc_'+ str(epoch)+'.h5')
        dec.save_weights('tf_vae/fashion/training_weights/dec_'+ str(epoch)+'.h5')
        print ('Time for epoch {0} is {1} sec'.format(epoch + 1, time.time()-start))

        # Generate after the final epoch
    display.clear_output(wait=True)
    generate_and_save_images([enc,sampling,dec],
                            epochs,
                            seed)
```

```python
# 추후 시각화를 위해 이미지 파일 저장해두기
def generate_and_save_images(model, epoch, test_input):
  # Notice `training` is set to False.
  # This is so all layers run in inference mode (batchnorm).
    m, v = enc(test_input, training=False)
    latent = sampling([m,  v])
    predictions = dec(latent, training=False)
    print(predictions.shape)
    fig = plt.figure(figsize=(4,4))

    for i in range(predictions.shape[0]):
        plt.subplot(5, 5, i+1)
        plt.imshow(predictions[i, :, :, 0] * 255, cmap='gray')
        plt.axis('off')

    plt.savefig('tf_vae/fashion/images/image_at_epoch_{:d}.png'.format(epoch))
    plt.show()
```
